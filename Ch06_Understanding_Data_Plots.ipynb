## Pandas loading csv

```python
from matplotlib import pyplot as plt
from pandas import read_csv
filename = '/Users/albertasuncion_mac/Documents/SMU/Python/machine_learning_mastery_with_python/Data/pima-indians-diabetes.csv'
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class'] 
data = read_csv(url, names=names)
print(data.shape)
```

```python
data.skew()
```

## Histogram

```python
data.hist(figsize=(10,10))
pyplot.show()
```

## Density plot

```python
data.plot(kind='density', subplots=True, layout=(3,3), sharex=False, figsize=(10,10)) 
pyplot.show()
```

```python
data.plot(kind='box', subplots=True, layout=(3,3), sharex=False, sharey=False, figsize=(10,10))
plt.show()
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(data, figsize=(15, 15))
plt.show()
```
